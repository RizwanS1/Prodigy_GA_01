```python
import logging
logging.basicConfig(
        format="%(asctime)s — %(levelname)s — %(name)s — %(message)s",
        datefmt="%m/%d/%Y %H:%M:%S",
        level=logging.INFO
    )

from aitextgen import aitextgen
from aitextgen.colab import mount_gdrive, copy_file_from_gdrive
```

```python
!nvidia-smi
```

```python
ai = aitextgen(tf_gpt2="124M", to_gpu=True)
```

```python
mount_gdrive()
```

```python
file_name = "input.txt"
```

```python
copy_file_from_gdrive(file_name)
```

```python
ai.train(file_name,
         line_by_line=False,
         from_cache=False,
         num_steps=3000,
         generate_every=1000,
         save_every=1000,
         save_gdrive=False,
         learning_rate=1e-3,
         fp16=False,
         batch_size=1,
         )
```

```python
ai = aitextgen(model_folder="trained_model", to_gpu=True)
```

```python
ai.generate()
```

```python
num_files = 5

for _ in range(num_files):
  ai.generate_to_file(n=1000,
                     batch_size=50,
                     prompt="ROMEO:",
                     max_length=256,
                     temperature=1.0,
                     top_p=0.9)
```
